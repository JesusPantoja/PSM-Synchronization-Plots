```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
from matplotlib import rcParams
rcParams['font.size'] = 16


version1=124   #41 cel, 124 
version2=122   #21 cel, 122 
version3=113   #11 cel, 113
fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d')

version = str(version1)+str(version2)+str(version3)

archnp1 = "/home/jesus/218/VCIR/Datos5Rep/Datos/Valores/valoresPO" + str(version1) + ".npy"
archnp2 = "/home/jesus/218/VCIR/Datos5Rep/Datos/Valores/valoresPO" + str(version2) + ".npy"
archnp3 = "/home/jesus/218/VCIR/Datos5Rep/Datos/Valores/valoresPO" + str(version3) + ".npy"
labels =[] #'1','2','3']
ysticks = np.arange(1,4)
# Make data.
X1 = np.arange(1, 21, 1)
Y1 = np.arange(3, 4, 1)
X1, Y1 = np.meshgrid(X1, Y1)


X2 = np.arange(1, 11, 1)
Y2 = np.arange(2, 3, 1)
X2, Y2 = np.meshgrid(X2, Y2)

X3 = np.arange(1, 6, 1)
Y3 = np.arange(1, 2, 1)
X3, Y3 = np.meshgrid(X3, Y3)
Z1=np.zeros([1,20])
Z2=np.zeros([1,10])
Z3=np.zeros([1,5])
ZZ=np.zeros([1,20])

valnp1 = np.load(archnp1)
valnp2 = np.load(archnp2)
valnp3 = np.load(archnp3)

Z1[0,:] = np.mean(valnp1[:,:,0],0)
Z2[0,:] = np.mean(valnp2[:,:,0],0)
Z3[0,:] = np.mean(valnp3[:,:,0],0)


ax.plot_wireframe(X1, Y1, Z1, color='blue', rstride=10, cstride=10, label="41 cells network")
ax.plot_wireframe(X2, Y2, Z2, color='blue', rstride=10, cstride=10, label="21 cells network")
ax.plot_wireframe(X3, Y3, Z3, color='blue', rstride=20, cstride=10, label="11 cells network")
ax.plot3D(X1[0,:],Y1[0,:],Z1[0,:], '-o', ms = 2, color='blue') 
ax.plot3D(X2[0,:],Y2[0,:],Z2[0,:], '-o', ms = 2, color = 'blue') 
ax.plot3D(X3[0,:],Y3[0,:],Z3[0,:], '-o', ms= 2, color = 'blue') 

### este codigo es para graficar las barras de error
err1 = np.std(valnp1, 0)/np.sqrt(valnp1.shape[0])
err2 = np.std(valnp2, 0)/np.sqrt(valnp2.shape[0])
err3 = np.std(valnp3, 0)/np.sqrt(valnp3.shape[0])

for i1 in range(X1.shape[1]):
    ax.plot3D([X1[0,i1],X1[0,i1]],[Y1[0,i1],Y1[0,i1]],[Z1[0,i1]-err1[i1,0],Z1[0,i1]+err1[i1,0]], '-_',color='blue') 

for i2 in range(X2.shape[1]):
    ax.plot3D([X2[0,i2],X2[0,i2]],[Y2[0,i2],Y2[0,i2]],[Z2[0,i2]-err2[i2,0],Z2[0,i2]+err2[i2,0]], '-_',color='blue') 

for i3 in range(X3.shape[1]):
    ax.plot3D([X3[0,i3],X3[0,i3]],[Y3[0,i3],Y3[0,i3]],[Z3[0,i3]-err3[i3,0],Z3[0,i3]+err3[i3,0]], '-_',color='blue') 

#####Hasta aqui es el codigo para graficar las barras de error


ax.set_zlim(0, 1.0)
ax.zaxis.set_major_locator(LinearLocator(6))
#ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
plt.yticks(ysticks, labels)
#ax.set_xticklabels(['A','B','C'])
ax.set_ylabel("Network Size", labelpad=15)
ax.set_zlabel("R", labelpad=15)
ax.set_xlabel("Connectivity Level", labelpad=10)
locator = matplotlib.ticker.MaxNLocator(nbins=5)
ax.xaxis.set_major_locator(locator)
ax.set_yticklabels(['11','21','41'])
ax.set_xticklabels(['1','5','10','15','20'])
plt.savefig("fig3DVCI.pdf")

ax.grid(False)
ax.xaxis.pane.set_edgecolor('black')
ax.yaxis.pane.set_edgecolor('black')
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False


view_1 = (10, -105)  #(25, -135)
view_2 = (25, -115)  #(25, -45)
init_view = view_1
ax.view_init(*init_view)

plt.show()

```
